```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('display.float_format', lambda x: '%.3f' % x)
sns.set_palette('Set2')  
```

```python
most_rated = pd.read_csv("data/most_rated/000000_0", sep='\t', header=None)
top_rated = pd.read_csv("data/top_rated/000000_0", sep='\t', header=None)
total_rating = pd.read_csv("data/total_rating/000000_0", sep='\t', header=None)
genre_rating = pd.read_csv("data/genre_rating/000000_0", sep='\t', header=None)
rating_dist = pd.read_csv("data/rating_dist/000000_0", sep='\t', header=None)
```

```python
rating_dist.sort_values(by=0, inplace=True)
```

```python
genre_rating.drop(genre_rating.tail(1).index,inplace=True)
genre_rating[1] = genre_rating[1].astype(float)
```

```python
ax = sns.barplot(x=1, y=0, data=most_rated)
ax.set(xlabel='No. Ratings', ylabel='Books')
# plt.xlim(0, most_rated[1].max())
plt.show()
```

```python
ax = sns.barplot(x=1, y=0, data=top_rated)
ax.set(xlabel='Rating', ylabel='Books')
plt.xlim(4.5, 4.8)
plt.show()
```

```python
ax = sns.barplot(x=1, y=0, data=total_rating)
ax.set(xlabel='Total Rating', ylabel='Books')
# plt.xlim(4.5, 4.8)
plt.show()
```

```python
ax = sns.barplot(x=1, y=0, data=genre_rating)
ax.set(xlabel='Rating', ylabel='Genre')
plt.xlim(3, 4.4)
plt.show()
```

```python
ax = sns.barplot(x=0, y=1, data=rating_dist)
ax.set(xlabel='Distribution', ylabel='No.Ratings')
# plt.xlim(3, 4.4)
plt.show()
```
