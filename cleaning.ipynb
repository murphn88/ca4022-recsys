```python
import pandas as pd
import numpy as np
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
children = pd.read_csv("data/children.csv", index_col=0)
comics = pd.read_csv("data/comics.csv", index_col=0)
fantasy = pd.read_csv("data/fantasy.csv", index_col=0)
history = pd.read_csv("data/history.csv", index_col=0)
mystery = pd.read_csv("data/mystery.csv", index_col=0)
poetry = pd.read_csv("data/poetry.csv", index_col=0)
romance = pd.read_csv("data/romance.csv", index_col=0)
ya = pd.read_csv("data/ya.csv", index_col=0)
```

```python
books = pd.concat([children, comics, fantasy, history, mystery, poetry, romance, ya])
books.head()
```

```python
eng = [x for x in books.language_code.unique() if 'en' in str(x) and str(x) != 'ben']
```

```python
books = books[books.language_code.isin(eng)]
```

```python
interactions = pd.read_csv("goodreads_interactions.csvgiiy081wtmp")
interactions
```

```python
book_lst = books.book_id.unique()

interactions = interactions[interactions.book_id.isin(book_lst)]
interactions
```

```python
interactions = interactions[(interactions['is_read'] != 0) & (interactions['rating'] != 0)]
interactions
```

```python
# interactions.groupby('user_id').count()
interactions['user_count'] = interactions.groupby('user_id')['user_id'].transform('count')
interactions.describe()
```

```python
interactions = interactions[interactions['user_count'] > 6]
interactions
```

```python
int_list = interactions.book_id.unique()
books = books[books.book_id.isin(int_list)]
books
```

```python
interactions.to_csv('data/interactions.csv')
books.to_csv('data/books.csv')
```
